```python
import pandas as pd
from datetime import datetime

# Загрузка старого CSV файла (если есть)
try:
    old_df = pd.read_csv('dancer_role_info.csv')
except FileNotFoundError:
    old_df = pd.DataFrame()
```

```python
new_df = pd.read_csv('dancer_role_info.csv')
```

```python
# Загрузка новых данных
new_df.loc[new_df['dancer_id'] == 1, 'primary_role_down_class'] = 'Adavanced'
new_df.loc[new_df['dancer_id'] == 2, 'primary_role_down_class'] = 'Champion'
# # Добавление столбца с датой формирования версии к старому DataFrame
# if not old_df.empty:
#     old_df['Version_Date'] = datetime.now().strftime("%Y-%m-%d %H:%M:%S")

# Добавление столбца с датой формирования версии к новому DataFrame
new_df['update_date'] = datetime.now().strftime("%Y-%m-%d")
```

```python
# Объединение старого и нового DataFrame
combined_df = pd.concat([old_df, new_df], ignore_index=True)
```

```python
# Удаление дубликатов (исключая столбец с датой формирования версии)
combined_df = combined_df.drop_duplicates(subset=combined_df.columns.difference(['update_date']), keep=False)
# Сортировка по dancer_id
combined_df = combined_df.sort_values(by='dancer_id')
```
